```python
from google.cloud import storage
import os

# Configurer les chemins
dataset_local_path = 'dataset'  # Le chemin local vers votre dossier dataset
bucket_name = 'dataset_reduit'  # Nom du bucket GCS

# Initialiser le client GCS
client = storage.Client()
bucket = client.bucket(bucket_name)

def upload_to_bucket(local_path, bucket, blob_name):
    """Uploader un fichier vers GCS."""
    blob = bucket.blob(blob_name)
    blob.upload_from_filename(local_path)
    print(f"Fichier {local_path} uploadé vers {blob_name}.")

def upload_directory_to_bucket(directory_path, bucket):
    """Uploader un répertoire et son contenu vers GCS."""
    for root, dirs, files in os.walk(directory_path):
        for file in files:
            local_file_path = os.path.join(root, file)
            relative_path = os.path.relpath(local_file_path, directory_path)
            blob_name = relative_path.replace(os.sep, '/')
            upload_to_bucket(local_file_path, bucket, blob_name)

# Uploader le répertoire 'dataset' vers le bucket GCS
upload_directory_to_bucket(dataset_local_path, bucket)

print("Le dataset a été uploadé avec succès.")
```
